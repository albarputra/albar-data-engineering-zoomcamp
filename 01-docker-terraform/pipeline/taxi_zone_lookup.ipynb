```python
engine = create_engine('postgresql://root:root@192.168.111.134:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='taxi_zone_lookup', con=engine))
```

```python
df.head(n=0).to_sql(name='taxi_zone_lookup', con=engine, if_exists='replace')
```

```python
df_iter = pd.read_csv(
        '../homework/taxi_zone_lookup.csv', iterator=True, chunksize=100000
    )
```

```python
first_chunk = next(df_iter)

first_chunk.head(0).to_sql(
    name="taxi_zone_lookup",
    con=engine,
    if_exists="replace"
)

print("Table created")

first_chunk.to_sql(
    name="taxi_zone_lookup",
    con=engine,
    if_exists="append"
)

print("Inserted first chunk:", len(first_chunk))

for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(
        name="taxi_zone_lookup",
        con=engine,
        if_exists="append"
    )
    print("Inserted chunk:", len(df_chunk))
```
